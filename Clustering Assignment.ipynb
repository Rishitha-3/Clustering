```python
#K-means ---->Elbow curve for different K values
#Elbow method is used to find the optimal number of clusters
```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Select the important columns
selected_columns = ['Balance', 'Qual_miles','cc1_miles' ,'cc2_miles' ,'cc3_miles' , 'Bonus_miles', 'Bonus_trans', 'Flight_miles_12mo', 'Flight_trans_12', 'Days_since_enroll','Award?']

# Scale the data
#scaler = StandardScaler()
#data_selected_scaled = scaler.fit_transform(data[selected_columns])

# Elbow method to find the optimal number of clusters
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, random_state=42)
    kmeans.fit(df_nom)
    wcss.append(kmeans.inertia_)

# Plot the Elbow curve
plt.figure(figsize=(10, 8))
plt.plot(range(1, 11), wcss, marker='o')
plt.title('Elbow Method for Optimal K')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
# Perform hierarchical clustering with different linkage criteria
from scipy.cluster.hierarchy import dendrogram, linkage
import matplotlib.pyplot as plt

# Perform hierarchical clustering with different linkage criteria
linkage_methods = ['ward', 'complete', 'average', 'single']

plt.figure(figsize=(15, 10))
for i, method in enumerate(linkage_methods):
    plt.subplot(2, 2, i+1)
    Z = linkage(df_nom, method=method)
    dendrogram(Z)
    plt.title(f'Hierarchical Clustering Dendrogram ({method.capitalize()} Linkage)')
    plt.xlabel('Sample Index')
    plt.ylabel('Distance')

plt.tight_layout()
plt.show()
```

```python
from sklearn.cluster import DBSCAN
import numpy as np

# Function to plot DBSCAN results
def plot_dbscan(data, eps, min_samples):
    dbscan = DBSCAN(eps=eps, min_samples=min_samples)
    labels = dbscan.fit_predict(data)
    
    plt.figure(figsize=(10, 8))
    plt.scatter(data[:, 0], data[:, 1], c=labels, cmap='viridis')
    plt.title(f'DBSCAN Clustering (eps={eps}, min_samples={min_samples})')
    plt.xlabel('Feature 1')
    plt.ylabel('Feature 2')
    plt.show()

```

```python
# Experiment with different epsilon and minPts values
eps_values = [0.3, 0.5, 0.7]
min_samples_values = [5, 10, 15]

# Apply PCA for visualization (optional)
pca = PCA(n_components=2)
data_pca = pca.fit_transform(df_nom)

for eps in eps_values:
    for min_samples in min_samples_values:
        plot_dbscan(data_pca, eps, min_samples)
```

```python
from sklearn.metrics import silhouette_score

# Optimal number of clusters from Elbow method
optimal_clusters = 3  # Example, replace with the actual optimal number from the Elbow method

kmeans = KMeans(n_clusters=optimal_clusters, random_state=42)
kmeans_labels = kmeans.fit_predict(df_nom)
silhouette_kmeans = silhouette_score(df_nom, kmeans_labels)
print(f'Silhouette Score for K-Means: {silhouette_kmeans}')
```

```python
from scipy.cluster.hierarchy import fcluster

# Using 'ward' linkage as an example
Z = linkage(df_nom, method='ward')
hierarchical_labels = fcluster(Z, t=optimal_clusters, criterion='maxclust')
silhouette_hierarchical = silhouette_score(df_nom, hierarchical_labels)
print(f'Silhouette Score for Hierarchical Clustering: {silhouette_hierarchical}')
```

```python
eps = 0.5
min_samples = 5

dbscan = DBSCAN(eps=eps, min_samples=min_samples)
dbscan_labels = dbscan.fit_predict(df_nom)
# Filter out noise points for silhouette score calculation
core_samples_mask = dbscan_labels != -1
silhouette_dbscan = silhouette_score(df_nom[core_samples_mask], dbscan_labels[core_samples_mask])
print(f'Silhouette Score for DBSCAN: {silhouette_dbscan}')
```
